### Word Embedding Techniques using Embedding Layer in Keras

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
# Vocabulary size
voc_size=5000
```

#### One Hot Representation

```python
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr) #out of the 5000 words, 4876 has the word 'the'
```

### Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# pre padding
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
# 10 feature dimesnions
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
#'the glass of milk',
embedded_docs[0]
```

```python
model.predict(embedded_docs[0])
```

```python
print(model.predict(embedded_docs))
```

```python
embedded_docs[0]
```

```python
print(model.predict(embedded_docs)[0])
```

```python
#Other sentences
new =["The world is a better place",
      "Marvel series is my favourite movie",
      "I like DC movies",
      "the cat is eating the food",
      "Tom and Jerry is my favourite movie",
      "Python is my favourite programming language"
      ]
```

```python
one_hot_repr=[one_hot(words,voc_size)for words in new]
print(one_hot_repr) #out of the 5000 words, 4876 has the word 'the'
```

```python
# post padding
new_length=10
embedded_docs=pad_sequences(one_hot_repr,padding='post',maxlen=new_length)
print(embedded_docs)
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=new_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
embedded_docs[0]
```

```python
model.predict(embedded_docs[0])
```

```python
print(model.predict(embedded_docs))
```
